```python
import pandas as pd
df = pd.read_csv("./raw_data/sales_subset.csv")
df["date"] = pd.to_datetime(df["date"])
df.head(2)
df.info()
```

```python
# mendapatkan bulan (angka) dari column date
df["month"] = df["date"].dt.month
df.head(2)
```

```python
# mendapatkan bulan (text) dari column date
df["month_name"] = df["date"].dt.month_name()
df.head(2)
```

```python
# mendapatkan weekday (angka) dari column date
# 0 --> Monday, 6 --> Sunday
df["day_of_week"] = df["date"].dt.weekday
df.head(2)
```

```python
# mendapatkan weekday (text) dari column date

df["day_of_week_text"] = df["date"].dt.day_name()
df.head(2)
```

```python
# mendapatkan tanggal hari dalam 1 bulan dari column date
df["day_of_month"] = df["date"].dt.day
df.head(2)
```

```python
# mendapatkan week dalam 1 tahun dari column date
df["week_of_year"] = df["date"].dt.isocalendar().week
df.head(2)
```

```python
# mendapatkan hari dalam 1 tahun dari column date
df["day_of_year"] = df["date"].dt.day_of_year
df.head(2)
```

```python
# mendapatkan  tahun dari column date
df["year"] = df["date"].dt.year
df.head(2)
```

```python
df.drop("unemployment", axis = 1, inplace = True)
df.head(2)
```

```python
from datetime import date
df["current_date"] = pd.to_datetime(date.today())
df.head(2)
```

```python
import numpy as np
today_date = date.today()
df["year_to_today"] = ((df["current_date"] - df["date"]) // np.timedelta64(1,"W"))
df.head(2)
```

```python
df.store.unique()
df_store_sales = df.groupby(["store", "year"], as_index=False)["weekly_sales"].sum()
df_store_sales
```

```python
df.store.unique()
df_store_sales = df.groupby(["type", "year"], as_index=False).agg({"weekly_sales" : "mean",
                                                                    "fuel_price_usd_per_l": "mean",
                                                                    "temperature_c": "mean"})
df_store_sales
```

```python
df_store_sales.rename(columns={"weekly_sales" : "sale_AVG", 
                               "fuel_price_usd_per_l" : "fuel_AVG", 
                               "temperature_c": "temp_AVG"}, inplace = True)
```

```python
df_pivot = df.pivot_table(columns=["year", "type"], index="month", values="weekly_sales", fill_value=0, aggfunc="mean")
df_pivot
```

```python
df_merge = df.merge(df_store_sales, on=["type", "year"], how="left")
df_merge
```

```python
df_pivot.to_csv("report_weekly_sales_based_on_month.csv")
```
